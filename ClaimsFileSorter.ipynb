```python
import os, shutil
```

```python
path = r"C:/Users/liamb/OneDrive/Documents/PythonClaimsSorter/"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['Word Files', 'Images', 'txt Files','CSV or Excel Files']

for loop in range(0,4):
   if not os.path.exists(path + folder_names[loop]):
       print(path + folder_names[loop])
       os.makedirs((path + folder_names[loop]))

for file in file_name:
    if ".csv" in file and not os.path.exists(path + "CSV or Excel Files/" + file):
        shutil.move(path + file, path + "CSV or Excel Files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "Images/" + file):
        shutil.move(path + file, path + "Images/" + file)
    elif ".txt" in file and not os.path.exists(path + "txt Files/" + file):
        shutil.move(path + file, path + "txt Files/" + file)
    elif ".xlsx" in file and not os.path.exists(path + "CSV or Excel Files/" + file):
        shutil.move(path + file, path + "CSV or Excel Files/" + file)
    elif ".docx" in file and not os.path.exists(path + "Word Files/" + file):
        shutil.move(path + file, path + "Word Files/" + file)
```
