```python
#import libraries

from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
#connecting to site

URL = "https://www.insurance.ca.gov/0250-insurers/0800-rate-filings/0100-rate-filing-lists/public-notices/index.cfm"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36"}
response = requests.get(URL, headers=headers)
soup1 = BeautifulSoup(response.content, "html.parser")
links = soup1.find_all("a", href =lambda x: x and x.endswith(".xlsx"))

for link in links:
    fileURL = "https://www.insurance.ca.gov" + link["href"]
    fileresponse = requests.get(fileURL, headers=headers, stream=True)
    filename = link.text.strip() + ".xlsx"
    with open(filename, "wb") as f:
        for chunk in fileresponse.iter_content(chunk_size=8192):
            f.write(chunk)
    time.sleep(1)
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
title = soup2.find('h1').get_text()
print(title)
```

```python
#cleaning title
title = title.strip()

print(title)
```
