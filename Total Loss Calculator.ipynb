```python
ClaimNumber = input("Please enter your claim number: ")

RepairEst = float(input("Enter repair estimate: $"))

ACV = float(input("Enter vehicles actual cash value: $"))

DDR = (RepairEst)/(ACV)

print(DDR)

if DDR>0:
    if(DDR<0.25):
        print("Your vehicle on Claim #" +ClaimNumber+ " has a damage degree of: MINOR.")
        print("Repairs should be straightforward.")
    elif(DDR<0.50):
        print("Your vehicle on Claim #" +ClaimNumber+ " has a damage degree of: MODERATE.")
        print("Repairs could take additional time, monitor for supplements from the shop.")
    elif(DDR<=0.75):
        print("Your vehicle on Claim #" +ClaimNumber+ " has a damage degree of: MAJOR.")
        print("Repairs may take a while. If you are using a rental, ensure it is active for length of repairs.")
    else:
        print("Your vehicle on Claim #" +ClaimNumber+ " is considered a TOTAL LOSS.")
        print("Your vehicle exceeds the total loss threshold of 75%- a Total Loss adjuster will continue working with you.")
else:
    print("Please provide a valid input.")
```

```python
#Damage Degree Ratio(DDR) = (Repair Estimate / Actual Cash Value)
```
